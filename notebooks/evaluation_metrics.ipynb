```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def read_and_plot(
        strategy: str,
        evaluation_metrics_path: str
        ):
    """
    Reads a CSV file and plots two metrics (loss and accuracy) in side-by-side subplots.

    Parameters:
    evaluation_metrics_path (str): Path to the evaluation metrics CSV file.

    The CSV should contain columns: 'round', 'eval_loss', and 'eval_accuracy'.
    """
    # Read the CSV file
    df = pd.read_csv(f"../{evaluation_metrics_path}")
    
    # Create the subplots
    fig, axes = plt.subplots(1, 2, figsize=(12, 5), constrained_layout=True)
    
    # Loss Plot
    sns.lineplot(ax=axes[0], data=df, x='round', y='eval_loss')
    axes[0].set_title('Evaluation Loss')
    axes[0].set_xlabel('Round')
    axes[0].set_ylabel('Evaluation Loss')
    
    # Accuracy Plot
    sns.lineplot(ax=axes[1], data=df, x='round', y='eval_accuracy')
    axes[1].set_title('Evaluation Accuracy')
    axes[1].set_xlabel('Round')
    axes[1].set_ylabel('Evaluation Accuracy')
    
    # Add a main title for the figure
    fig.suptitle(f'Evaluation Metrics using {strategy}', fontsize=16)
    
    # Display the plots
    plt.show()
```

## google-t5/t5-small/FedProx

Model: google-t5/t5-small  
Strategy: FedProx  
Date-Time: 22024-12-28/13-00-12

```python
read_and_plot(
    strategy="FedProx",
    evaluation_metrics_path="models/google-t5/t5-small/2024-12-28/13-00-12/evaluation_metrics.csv"
)
```
